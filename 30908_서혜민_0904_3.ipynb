```python
# exam에서 nclass가 1인 경우만 추출
exam.query('nclass == 1 ')
```

```python
# exam에서 nclass가 1반이 아닌 경우 추출
exam.query('nclass != 1')
```

```python
# 수학점수가 50점을 초과한 경우
exam.query('math>50')
```

```python
# 영어 점수가 80점 이하인 경우
exam.query('english<=80')
```

```python
#1반이면서 수학 점수가 50점 이상인 경우
exam.query('nclass == 1 & math>=50')
```

```python
#수학 점수가 90점이상이거나 영어 점수가 90점 이상인 경우
exam.query('math>=90 | english>=90')
```

```python
# 1,3,4반에 해당하는 반만 출력
exam.query('nclass != 2')
```
