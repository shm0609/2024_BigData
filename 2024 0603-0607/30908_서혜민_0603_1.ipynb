```python
#그꾸

#꺾은선 그래프 그리기
#라이브러리 불러오기 및 별명 지정하기
import matplotlib.pyplot as plt
#한글폰트 적용하기
plt.rc('font',family='Malgun Gothic')

#x,y축 데이터 지정
sales = [1,5,7,3,7]
month= ['mar','apr','may','jun','jul']

#그래프의 해상도 설정하기 dpi figure
plt.figure(dpi=100)

#제목 title
plt.title('월별 판매 실적')

#꺾은선 형 그래프 그리기
#그래프 범례 label
plt.plot(month,sales,color='r',label='단위:개수')

#범례 보이게 설정.lower right
plt.legend(loc='lower right')
#x축,y축 x label, y label
plt.xlabel('월')
plt.ylabel('매출')

#그래프 보이기
plt.show()
```
